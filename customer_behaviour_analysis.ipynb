**from that we get that review rating has 37 null values**

```python
df['Category'].unique()
```

**change the null values in review rating columns with the median of each category**

```python
df['Review Rating']=df['Review Rating'].fillna(df.groupby('Category')['Review Rating'].transform('median'))
```

**convert all columns name to snake casing and replace all space to _**

```python
df.columns=df.columns.str.lower().str.replace(' ','_')
```

```python
df.head()
```

**create a column age_group**

```python
labels=['Young_adult','Adult','Middle_Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels = labels)
```

```python
df[['age','age_group']].head()
```

**get the no of days customer placed an orders based on frequency of puchase**

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping ={
    'Fortnightly':14,
    'Weekly':7,
    'Annually': 365,
    'Quarterly':90,
    'Bi-Weekly': 15,
    'Monthly' : 30,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
